```python
from duckduckgo_search import DDGS
import httpx
from bs4 import BeautifulSoup
```

```python
def search(query, max_results=3):
    r = httpx.get("https://searx.be/search", params={
        "q": query, "format": "json", "engines": "google,bing"
    })
    return r.json()["results"][:max_results]
```

```python
with DDGS() as ddgs:
    results = list(ddgs.text("Iran Côte d'Ivoire strike attack 2025", max_results=3))
    results.extend(list(ddgs.text("Abidjan bombing June 2025", max_results=3)))
    results.extend(list(ddgs.text("Iran military strikes multiple countries 2025", max_results=3)))
    results.extend(list(ddgs.text("How many countries has Iran attacked 2025", max_results=3)))

print(results)
for r in results:
    print(r['href'])  
    print(r['title'])  
    print(r['body'])  
```
